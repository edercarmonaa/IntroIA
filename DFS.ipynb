# Búsqueda primero en profundidad (DFS)
Este algoritmo se centra en recorrer los nodos desde la raíz e ir expandiendo cada uno de los nodos que va localizando, de forma recurrente (desde el nodo padre hacia el nodo hijo), en el caso de llegar a un "callejón sin salida" se retrocede hasta el nodo más cercano desde donde se pueda tomar una rama alternativa para poder seguir avanzando.
